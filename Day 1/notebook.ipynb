# NumPy Practice Exercises

### Question 1
Create a 1D NumPy array containing the numbers from 10 to 50 (inclusive) with a step of 5.

```python
import numpy as np
arr1 = np.arange(10, 51, 5)
print(arr1)
```

### Question 2
Create a 4x6 array filled with zeros, then fill its third row (index 2) with the value 99.

```python
arr2 = np.zeros((4, 6))
arr2[2, :] = 99
print(arr2)
```

### Question 3
Given the array: `arr = np.array([12, 45, 7, 19, 88, 3, 56, 91, 24, 67])`, print all elements that are greater than 50.

```python
arr = np.array([12, 45, 7, 19, 88, 3, 56, 91, 24, 67])
print(arr[arr > 50])
```

### Question 4
Create a 5x5 identity matrix, then change its main diagonal (from top-left to bottom-right) to the values [10, 20, 30, 40, 50].

```python
arr4 = np.eye(5)
np.fill_diagonal(arr4, [10, 20, 30, 40, 50])
print(arr4)
```

### Question 5
Let `x = np.arange(1, 21).reshape(4, 5)`. Extract the following using slicing:

a. The last two rows

b. The third column

c. The 3x3 submatrix in the top-right corner

```python
x = np.arange(1, 21).reshape(4, 5)
print("a. Last two rows:\n", x[-2:, :])
print("b. Third column:\n", x[:, 2])
print("c. Top-right 3x3:\n", x[:3, 2:])
```

### Question 6
Given two arrays: `a = np.array([1, 2, 3, 4])` and `b = np.array([[10], [20], [30]])`.

Compute `a + b` and explain the shape.

```python
a = np.array([1, 2, 3, 4])
b = np.array([[10], [20], [30]])
result = a + b
print("Resulting Array:\n", result)
print("Shape:", result.shape)
```

### Question 7
Normalize the following array to the range [0, 1]:
`data = np.array([150, 220, 90, 300, 180, 45])`

```python
data = np.array([150, 220, 90, 300, 180, 45])
normalized = (data - data.min()) / (data.max() - data.min())
print(normalized)
```

### Question 8
Standardize (z-score) the same array from question 7.

```python
standardized = (data - data.mean()) / data.std()
print(standardized)
```

### Question 9
Given the 2D array `mat`, compute row means, column sums, and max value position.

```python
mat = np.array([[4, 8, 1, 12], [7, 3, 9, 5], [11, 2, 6, 10], [15, 0, 14, 13]])
print("Row Means:", mat.mean(axis=1))
print("Col Sums:", mat.sum(axis=0))
max_idx = np.unravel_index(mat.argmax(), mat.shape)
print(f"Max Value: {mat.max()} at {max_idx}")
```

### Question 10

### Question 11
Reverse the order of rows and columns for a 4x4 array.

```python
arr11 = np.arange(1, 17).reshape(4, 4)
print(arr11[::-1, ::-1])
```

### Question 12 & 13
Filtering names by age and replacing values.

```python
ages = np.array([23, 45, 19, 34, 28, 51, 17, 39])
names = np.array(["Ali", "Sara", "Omar", "Lina", "Khaled", "Nour", "Yara", "Hassan"])
print("Names (25-40):", names[(ages >= 25) & (ages <= 40)])
ages[ages < 20] = 20
print("Updated ages:", ages)
```

### Question 14

### Question 15
10 random numbers from normal distribution (Mean=100, SD=15).

### Question 16 & 17
Random selection and shuffling.

```python
colors = np.array(["red", "blue", "green", "yellow", "purple"])
print("Random colors:", np.random.choice(colors, 3, replace=False))
nums = np.arange(1, 21)
np.random.shuffle(nums)
print("Shuffled 1-20:", nums)
```
