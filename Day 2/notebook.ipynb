### Using .loc, change Khaled's grade from 55 to 68.

```python
data.loc[data['Name'] == 'Khaled', 'Grade'] = 68
```

```python
data[data['Name'] == 'Khaled']
```

### Add a new column Is_Excellent that is True if Grade ≥ 90, False otherwise (vectorized)

```python
data['Is_Excellent'] = data['Grade'] >= 90
```

```python
data
```

### Fill missing Grade values with the average Grade.

```python
data['Grade'].fillna(data['Grade'].mean(skipna=True), inplace=True)
```

```python
data
```

### Fill missing Score Date values with the most common date (mode).

```python
mode_value = data['Score Date'].mode()[0]

data['Score Date'] = data['Score Date'].fillna(mode_value)
```

```python
data
```

### Create a column Grade_Category using pd.cut with bins.

```python
data['Grade_Category'] = pd.cut(data['Grade'], bins=[0, 69, 79, 89, 100], labels=["Fail", "Average", "Good", "Excellent"])
```

```python
data
```

### Remove the column Student ID.

```python
data.drop(columns="Student ID", inplace=True)
```

```python
data
```

### Keep only students who Passed.

```python
data_passed = data[data['Grade_Category'] != "Fail"]
```

```python
data_passed
```

### Sort the DataFrame by Grade descending.

```python
data_passed.sort_values(by="Grade", ascending=False, inplace=True)
data_passed
```

```python
data_passed = data_passed.sort_values(by=['City', 'Grade'], ascending=[True, False])
data_passed
```

### Add a column Rank that ranks students by Grade

```python
data_passed['Rank'] = data['Grade'].rank(ascending=False, method='min').astype(int)
```

```python
data_passed
```

### Show the top 4 students with the highest grades.

```python
data_passed.nlargest(columns='Grade', n=4)
```

### Show the 3 youngest students.

```python
data_passed.nsmallest(columns='Age', n=3)
```

### Find the student(s) with the highest Grade and show their Name and Grade.

```python
max_grade = data['Grade'].max()

data_passed[data_passed['Grade'] == max_grade]
```

### After filling missing grades, recalculate the average Grade per Subject.

```python
print("Average Grade:", data['Grade'].mean(), end="")
```

### Create a new column City_Avg that shows the average grade of each student's city.

```python
data_passed['City_Avg'] = data_passed.groupby('City')['Grade'].transform('mean')
```

```python
data_passed
```

### Show the subject with the highest average grade.

```python
subject_averages = data.groupby('Subject')['Grade'].mean()

best_subject = subject_averages.idxmax()
highest_score = subject_averages.max()

print(f"The highest performing subject is {best_subject} with an average of {highest_score:.2f}")
```
